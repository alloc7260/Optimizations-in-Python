```python
from tqdm.notebook import tqdm
import time
```

# Regular Looping

## Regular For Loop

#### Pros :
* Execute in Sequence
* Support mult-iliner

#### Cons :
* Time Consuming

```python
for i in tqdm(range(20)):
    #code
    print(i)
    time.sleep(1)
```

## List Comprehension

#### Pros :
* Slightly Time Efficient
* Execute in Sequence


#### Cons :
* only one-liners
* Support mult-iliner with function execution

```python
def fuc(i):
  print(i)
  time.sleep(1)

[fuc(i) for i in tqdm(range(20))]
```

## Multiprocessed For Loop

#### Pros :
* Time Efficient
* Support mult-iliner

#### Cons :
* Execute in Random Sequence

```python
from multiprocessing import Pool

def fuc(i):
    # code
    print(i)
    time.sleep(1)

npools = 20 # define number of pools as per machine configuration (higher the better)

with Pool(npools) as pp:
  pp.map(fuc,tqdm(range(20)))
```

# Extract or Store Value

## Regular For Loop

#### Pros :
* Execute in Sequence
* Support mult-iliner

#### Cons :
* Time Consuming

```python
%%time
output = []
for i in tqdm(range(20)):
    #code
    output.append(i)
    time.sleep(1)
output
```

## List Comprehension

#### Pros :
* Slightly Time Efficient
* Execute in Sequence


#### Cons :
* only one-liners
* Support mult-iliner with function execution

```python
%%time

def fuc(i):
  time.sleep(1)
  return i

output = [fuc(i) for i in tqdm(range(20))]
output
```

## Multiprocessed For Loop

#### Pros :
* Time Efficient
* Support mult-iliner

#### Cons :
* Execute in Random Sequence

```python
%%time
from multiprocessing import Pool

def fuc(i):
    # code
    time.sleep(1)
    return i

npools = 20 # define number of pools as per machine configuration (higher the better)

with Pool(npools) as pp:
  output = pp.map(fuc,tqdm(range(20)))

output
```

for processes that take longer time and can be execute in parallel are good with multiprocessing

⚠ Disclaimer : applying multiprocessing in every loop may lead to inefficiency in program
